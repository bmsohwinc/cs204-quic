```python
import pandas as pd
import matplotlib.pyplot as plt
from analyze import QLogAnalyzer
```

```python
qla = QLogAnalyzer("../pkt_logs/client_20251109-125125_log.json")
```

```python
df = pd.DataFrame(qla.analyze_cwnd())

# Convert large timestamps to relative (recommended)
t0 = df["time"].iloc[0]
df["time_rel"] = df["time"] - t0

# Plot
plt.figure(figsize=(10, 5))

plt.plot(df["time_rel"], df["cwnd"], marker="o", label="cwnd")
# plt.plot(df["time_rel"], df["bytes_in_flight"], marker="o", label="bytes_in_flight")

plt.xlabel("Time (relative)")
plt.ylabel("Value")
plt.title("CWND & Bytes-in-Flight Over Time")
plt.legend()
plt.grid(True)

plt.show()
```
